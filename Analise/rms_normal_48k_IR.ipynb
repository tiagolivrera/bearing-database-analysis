```python
# import das bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sc
```

```python
## carregamento dos dados
# ------ Normal Baseline ------
nb_1 = sc.loadmat('../Normal/97.mat')
nb_2 = sc.loadmat('../Normal/98.mat')
nb_3 = sc.loadmat('../Normal/99.mat')
nb_4 = sc.loadmat('../Normal/100.mat')
# 48k Drive End Bearing Fault Data, 0.007', Inner Race
de_1 = sc.loadmat('../48k Drive End Bearing Fault Data/Inner Race/IR007_0.mat')
de_2 = sc.loadmat('../48k Drive End Bearing Fault Data/Inner Race/IR007_1.mat')
de_3 = sc.loadmat('../48k Drive End Bearing Fault Data/Inner Race/IR007_2.mat')
de_4 = sc.loadmat('../48k Drive End Bearing Fault Data/Inner Race/IR007_3.mat')
```

```python
de_4['X112_DE_time']
```

```python
time_step = 1/48000.0
velocity_de = np.zeros_like(de_4['X112_DE_time'])
velocity_fe = np.zeros_like(de_4['X112_FE_time'])
for i in range(1, len(de_4['X112_DE_time'])):
    velocity_de[i] = velocity_de[i-1] + de_4['X112_DE_time'][i] * time_step
    velocity_fe[i] = velocity_fe[i-1] + de_4['X112_FE_time'][i] * time_step
```

```python
rms_de = np.sqrt(np.mean(velocity_de**2))
print("rms_de: ", rms_de)
rms_fe = np.sqrt(np.mean(velocity_fe**2))
print("rms_fe: ", rms_fe)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(velocity_de, label='Velocidade DE')
plt.plot(velocity_fe, label='Velocidade FE')
plt.axhline(y=rms_de, color='r', linestyle='--', label=f'RMS DE: {rms_de:.2f}')
plt.axhline(y=rms_fe, color='g', linestyle=':', label=f'RMS FE: {rms_fe:.2f}')
plt.xlabel('Sample')
plt.ylabel('Value')
plt.legend()
plt.title('Root Mean Square (RMS): 48k Bearing Fault, 3 HP Motor Load, Inner Race')
plt.grid(True)
plt.show()
```

```python
# Calculando as velocidades a partir da acelerao
time_step = 1/48000.0                                # Os dados so captados a uma taxa de 48000 samples/segundo
# Normal Baseline
velocity_de_nb_1 = np.zeros_like(nb_1['X097_DE_time'])
velocity_fe_nb_1 = np.zeros_like(nb_1['X097_FE_time'])
velocity_de_nb_2 = np.zeros_like(nb_2['X098_DE_time'])
velocity_fe_nb_2 = np.zeros_like(nb_2['X098_FE_time'])
velocity_de_nb_3 = np.zeros_like(nb_3['X099_DE_time'])
velocity_fe_nb_3 = np.zeros_like(nb_3['X099_FE_time'])
velocity_de_nb_4 = np.zeros_like(nb_4['X100_DE_time'])
velocity_fe_nb_4 = np.zeros_like(nb_4['X100_FE_time'])
# 48k Bearing Fault
velocity_de_1 = np.zeros_like(de_1['X109_DE_time']) 
velocity_fe_1 = np.zeros_like(de_1['X109_FE_time'])
velocity_de_2 = np.zeros_like(de_2['X110_DE_time'])
velocity_fe_2 = np.zeros_like(de_2['X110_FE_time'])
velocity_de_3 = np.zeros_like(de_3['X111_DE_time'])
velocity_fe_3 = np.zeros_like(de_3['X111_FE_time'])
velocity_de_4 = np.zeros_like(de_4['X112_DE_time'])
velocity_fe_4 = np.zeros_like(de_4['X112_FE_time'])

for i in range(1, len(nb_1['X097_DE_time'])):
    velocity_de_nb_1[i] = velocity_de_nb_1[i-1] + nb_1['X097_DE_time'][i] * time_step
    velocity_fe_nb_1[i] = velocity_fe_nb_1[i-1] + nb_1['X097_FE_time'][i] * time_step
for i in range(1, len(nb_2['X098_DE_time'])):
    velocity_de_nb_2[i] = velocity_de_nb_2[i-1] + nb_2['X098_DE_time'][i] * time_step
    velocity_fe_nb_2[i] = velocity_fe_nb_2[i-1] + nb_2['X098_FE_time'][i] * time_step
for i in range(1, len(nb_3['X099_DE_time'])):
    velocity_de_nb_3[i] = velocity_de_nb_3[i-1] + nb_3['X099_DE_time'][i] * time_step
    velocity_fe_nb_3[i] = velocity_fe_nb_3[i-1] + nb_3['X099_FE_time'][i] * time_step
for i in range(1, len(nb_4['X100_DE_time'])):
    velocity_de_nb_4[i] = velocity_de_nb_4[i-1] + nb_4['X100_DE_time'][i] * time_step
    velocity_fe_nb_4[i] = velocity_fe_nb_4[i-1] + nb_4['X100_FE_time'][i] * time_step

for i in range(1, len(de_1['X109_DE_time'])):
    velocity_de_1[i] = velocity_de_1[i-1] + de_1['X109_DE_time'][i] * time_step
    velocity_fe_1[i] = velocity_fe_1[i-1] + de_1['X109_FE_time'][i] * time_step
for i in range(1, len(de_2['X110_DE_time'])):
    velocity_de_2[i] = velocity_de_2[i-1] + de_2['X110_DE_time'][i] * time_step
    velocity_fe_2[i] = velocity_fe_2[i-1] + de_2['X110_FE_time'][i] * time_step
for i in range(1, len(de_3['X111_DE_time'])):
    velocity_de_3[i] = velocity_de_3[i-1] + de_3['X111_DE_time'][i] * time_step
    velocity_fe_3[i] = velocity_fe_3[i-1] + de_3['X111_FE_time'][i] * time_step
for i in range(1, len(de_4['X112_DE_time'])):
    velocity_de_4[i] = velocity_de_4[i-1] + de_4['X112_DE_time'][i] * time_step
    velocity_fe_4[i] = velocity_fe_4[i-1] + de_4['X112_FE_time'][i] * time_step
```

```python
# Valores RMS
rms_de_nb_1 = np.sqrt(np.mean(velocity_de_nb_1**2))
rms_de_nb_2 = np.sqrt(np.mean(velocity_de_nb_2**2))
rms_de_nb_3 = np.sqrt(np.mean(velocity_de_nb_3**2))
rms_de_nb_4 = np.sqrt(np.mean(velocity_de_nb_4**2))
rms_fe_nb_1 = np.sqrt(np.mean(velocity_fe_nb_1**2))
rms_fe_nb_2 = np.sqrt(np.mean(velocity_fe_nb_2**2))
rms_fe_nb_3 = np.sqrt(np.mean(velocity_fe_nb_3**2))
rms_fe_nb_4 = np.sqrt(np.mean(velocity_fe_nb_4**2))

rms_de_1 = np.sqrt(np.mean(velocity_de_1**2))
rms_de_2 = np.sqrt(np.mean(velocity_de_2**2))
rms_de_3 = np.sqrt(np.mean(velocity_de_3**2))
rms_de_4 = np.sqrt(np.mean(velocity_de_4**2))
rms_fe_1 = np.sqrt(np.mean(velocity_fe_1**2))
rms_fe_2 = np.sqrt(np.mean(velocity_fe_2**2))
rms_fe_3 = np.sqrt(np.mean(velocity_fe_3**2))
rms_fe_4 = np.sqrt(np.mean(velocity_fe_4**2))
```
